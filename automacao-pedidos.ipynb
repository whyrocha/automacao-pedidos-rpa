```python
import pyautogui
import pandas as pd 
import time 
import cv2

pyautogui.alert(text='O Código vai começar, não mexa no PC', title='Pedido Mecânica', button='OK') 

time.sleep(2)

pyautogui.PAUSE = 2

#importar base de dados
tabela = pd.read_csv('pedidos.csv')
print(tabela)


#abrir sistema
x, y, largura, altura = pyautogui.locateOnScreen('abrir_sistema.png', confidence=0.8)
pyautogui.click(x + largura/2, y + altura/2)

#abrir cotação
time.sleep(1)
x, y, largura, altura = pyautogui.locateOnScreen('abrir_cotacao.png', confidence=0.8)
pyautogui.click(x + largura/2, y + altura/2)

#criar nova cotação
x, y, largura, altura = pyautogui.locateOnScreen('criar_cotacao.png', confidence=0.8)
pyautogui.click(x + largura/2, y + altura/2)

#adicionar nome do pedido
pyautogui.write("Nome do Pedido e Data")

#clica no campo codigos
pyautogui.click(x=359, y=332)

#preencher os codigos e quantidades
for linha in tabela.index:
    time.sleep(0.50)
    pyautogui.write(str(tabela.loc[linha, "codigo"]))
    pyautogui.press("tab")
    pyautogui.write(str(tabela.loc[linha, "qnt"]))
    pyautogui.press("f2")
    
    
pyautogui.alert(text='Pedido digitado, prossiga', title='Pedido Mecânica', button='OK') 
```
